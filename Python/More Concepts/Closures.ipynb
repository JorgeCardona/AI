# <center> Closures </center>

```python
def pows(x):
    
    "This is the enclosing function"
    def enclosing_function(y):
        
        "The nested function"
        return x ** y
    
    "return the function"
    return enclosing_function
```

```python
r = pows(2)
s = pows(4)
```

```python
r(3)
```

```python
s(5)
```

```python
r(s(7))
```

```python
def show_function(func): 
    
    def select_funtion(*args): 
        print('Running "{}" with arguments {}'.format(func.__name__, args)) 
        print(func(*args)) 
        
    # Necessary for closure to work (returning WITHOUT parenthesis) 
    return select_funtion               
```

```python
def sumar(x, y): 
    return x+y 

def restar(x, y): 
    return x-y

def multiplicar(x, y): 
    return x*y 
```

```python
sumar_show_function = show_function(sumar) 
restar_show_function = show_function(restar)
multiplicar_show_function = show_function(multiplicar)
```

```python
sumar_show_function(3, 4) 
sumar_show_function(5, 6)
```

```python
restar_show_function(2, 4) 
restar_show_function(6, 8)
```

```python
multiplicar_show_function(5, 10) 
multiplicar_show_function(15, 20)
```
