```python
methods = ['GET','POST','PUT','PATCH','DELETE','COPY','HEAD','OPTIONS','LINK','UNLINK','PURGE','LOCK','UNLOCK','PROPFIND','PROPPATCH','VIEW','CONNECT','TRACE','SEARCH','MOVE','MKCOL','MERGE','UPDATE','LABEL']
```

```python
def break_instruction(methods):
    
    instruction = 'BREAK'
    
    msg1 =  "starting instruction {}" .format(instruction)    
    msg3 =  "-------{}--------" .format(instruction)    
    msg5 =  "ending instruction {}" .format(instruction)
    
    
    print(msg1)
    
    for i in methods:

        if(len(i) %2 != 0):
            
            msg2 =  "this value comply {}" .format(i)
            
            print(msg2)
            break
            print(msg3)

        else:
            msg4 =  "this value does not comply with the restriction {}, {}".format(instruction,i)
            print(msg4)
    
    print(msg5)
    print()
```

```python
def pass_instruction(methods):
    
    instruction = 'PASS'
    
    msg1 =  "starting instruction {}" .format(instruction)    
    msg3 =  "-------{}--------" .format(instruction)    
    msg5 =  "ending instruction {}" .format(instruction)
    
    
    print(msg1)
    
    for i in methods:

        if(len(i) %2 != 0):
            
            msg2 =  "this value comply {}" .format(i)
            
            print(msg2)
            pass
            print(msg3)

        else:
            msg4 =  "this value does not comply with the restriction {}, {}".format(instruction,i)
            print(msg4)
    
    print(msg5)
    print()
```

```python
def continue_instruction(methods):
    
    instruction = 'CONTINUE'

    msg1 =  "starting instruction {}" .format(instruction)    
    msg3 =  "-------{}--------" .format(instruction)    
    msg5 =  "ending instruction {}" .format(instruction)
    
    
    print(msg1)
    
    for i in methods:

        if(len(i) %2 != 0):
            
            msg2 =  "this value comply {}" .format(i)
            
            print(msg2)
            continue
            print(msg3)

        else:
            msg4 =  "this value does not comply with the restriction {}, {}".format(instruction,i)
            print(msg4)
    
    print(msg5)
    print()
```

```python
def evaluate_instructions(methods):
    pass_instruction(methods)
    
    continue_instruction(methods)
    
    break_instruction(methods)     
```

```python
evaluate_instructions(methods)
```
