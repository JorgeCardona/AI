# <center> Genero una lista de 100 millones de números aleatorios</center>

```python
import numpy as np
x = list(np.random.randint(low=1,high=500000,size=99999999))
```

# <center> Compejidad Constante O(k) Constant Time </center>

```python
%%time
def constante(x:list) -> list:
    return x

constante(x)
```

# <center> Complejidad O(n) Linear Time </center>

```python
%%time
def iterador_normal(x:list) -> list:
    
    contador = len(x)
    
    while(contador > 0):
        contador -= 1
    
    return x

iterador_normal(x)
```

# <center> Complejidad O(n<sup>2</sup>)  Quadratic Time </center>

```python
%%time
def iterador_anidado(x:list) -> list:
    
    contador_externo = len(x)//1000
    contador_interno = len(x)//1000
    
    while(contador_externo > 0):
        contador_externo -= 1
        
        for i in range(contador_interno):
            i
    return x

iterador_anidado(x)
```

# <center> Complejidad O(log(n)) Logarithmic Time</center>

```python
%%time
def iterador_multiplicado(x:list) -> list:
    
    iterador = len(x)    
    incremento = 1
    while(iterador > 0):
        iterador -=  incremento
        incremento *= (incremento + 1)
    
    return x
        
iterador_multiplicado(x)
```

# <center> complejidad O(log(n)) Logarithmic Time</center>

```python
%%time
def iterador_dividido(x:list) -> list:    
    
    iterador = -len(x)    
    incremento = 1
    while(iterador < 0):
        iterador /=  incremento
        incremento += 1
    
    return x
        
iterador_dividido(x)
```

# <center> Complejidad O(log(log(n))) Logarithmic from Logarithmic Time</center>

```python
%%time

import math

def iterador_incremento_exponencial(x:list) -> list:    
    
    iterador = len(x)    
    incremento = 1
    while(iterador > 0):
        iterador -=  pow(incremento, 2)
        incremento += 1
    
    return x
        
iterador_incremento_exponencial(x)
```
